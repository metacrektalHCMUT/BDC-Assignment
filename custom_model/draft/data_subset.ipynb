```python
data_path = "/content/drive/MyDrive/Big_Data/custom_model/clean_data.csv"
```

```python
data_full = pd.read_csv(data_path,delimiter=',')
```

```python
neg_data = data_full[data_full["target"] == "NEGATIVE"]
pos_data = data_full[data_full["target"] == "POSITIVE"]
```

```python
data_full["text"] = data_full["text"].apply(str)
```

```python
neg_data = neg_data.sample(frac = 1)[:50000]
pos_data = pos_data.sample(frac = 1)[:50000]
```

```python
data = pd.concat([neg_data,pos_data])
```

```python
train_data_df, test_data_df = train_test_split(data, test_size=0.2, random_state=42)
```

SAVE_SUB_DATA
